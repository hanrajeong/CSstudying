# ndarray 데이터를 이용하여 다양한 그래프 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 #주피터 노트북 내에서 그래프를 그리기 위한 코드
```

```python
x = np.linspace(0, 10, 11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

```python
#Plot 함수

plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
# 그래프에 주석 추가
# x, y축 및 타이틀 추가

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y') #제목
plt.grid(True) #모눈선 추가
plt.xlim(0, 20) #그래프에 나타나는 x값 범위
plt.ylim(0,200) # y값 범위
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r') #색상을 지정, 'k'는 검은색. RGB 색을 입력할 수 있음. 스타일은 공식문서에서 찾아서 적용하면 됨.
```

```python
plt.plot(x, y, '-.')
```

```python
plt.plot(x, y, 'g^')
```

```python
plt.plot(x, y, 'm:')
```

```python
plt.plot(x, y, 'm:', linewidth=9)
```

https://matplotlib.org/2.0.2/api/pyplot_api.html
위의 주소에서 detail 한 사항은 확인가능함.

```python
plt.plot(x, y, color = 'black', linestyle='--', marker='^', markerfacecolor = 'blue', markersize = 6)
```

```python
#subplot : 구획을 나누어서 여러 함수를 그리는 것

plt.subplot(2, 2, 1) #앞에 2, 2는 2행 2열로 총 4개의 함수를 그리겠다는 것.
plt.plot(x, y, 'r')

plt.subplot(2,2,2)
plt.plot(x, y, 'g')

plt.subplot(2,2,3)
plt.plot(x, y, 'k')

plt.subplot(2,2,4)
plt.plot(x,y, 'b')
```

```python
#hist 함수 사용
#histogram 생성
#bins로 histogram bar 개수 설정

data = np.random.randint(1, 100, size = 200)
print(data)
```

```python
plt.hist(data, bins=10, alpha = 0.3) #alpha는 투명도, bins는 막대기 개수
plt.xlabel('values')
plt.ylabel('data')
plt.grid(True)
```
